### CUSTOMERS

```python
customers = [
(1,"John","Chennai"),
(2,"David","Bangalore"),
(3,"Rahul","Hyderabad"),
(4,"Priya","Mumbai"),
(5,"Anjali","Delhi"),
(6,"Kumar","Chennai"),
(7,"Ravi","Coimbatore"),
(8,"Sneha","Pune"),
(9,"Asha","Kolkata"),
(10,"Vijay","Chennai")
]

customer_df = spark.createDataFrame(
customers,
["customer_id","customer_name","city"]
)

customer_df.write.mode("overwrite").saveAsTable(
"cicd_dev.default.customers"
)
```
