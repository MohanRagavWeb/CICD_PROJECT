```python
def validate_row_count(table_name):

    count = spark.table(
        table_name
    ).count()

    if count == 0:
        raise Exception(
            f"{table_name} is empty"
        )

    print(
        f"{table_name} passed"
    )
```

```python
validate_row_count(
    "cicd_dev.gold.gold_customer_analytics"
)

validate_row_count(
    "cicd_dev.gold.gold_product_analytics"
)

validate_row_count(
    "cicd_dev.gold.gold_sales_analytics"
)

validate_row_count(
    "cicd_dev.gold.gold_inventory_analytics"
)
```
