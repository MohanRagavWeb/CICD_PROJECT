```python
from pyspark.sql.functions import *

def validate_table(
    source_table,
    target_table,
    key_column
):

    print(f"\nValidating {target_table}")

    source_df = spark.table(source_table)
    target_df = spark.table(target_table)

    # -------------------------
    # Row Count Validation
    # -------------------------

    source_count = source_df.count()
    target_count = target_df.count()

    print(f"Source Count : {source_count}")
    print(f"Target Count : {target_count}")

    if source_count != target_count:
        raise Exception(
            f"Row Count Mismatch : {target_table}"
        )

    # -------------------------
    # Schema Validation
    # -------------------------

    source_schema = source_df.columns
    target_schema = target_df.columns

    if source_schema != target_schema:
        raise Exception(
            f"Schema Mismatch : {target_table}"
        )

    # -------------------------
    # Null Validation
    # -------------------------

    null_count = target_df.filter(
        col(key_column).isNull()
    ).count()

    if null_count > 0:
        raise Exception(
            f"Null Keys Found : {target_table}"
        )

    print(
        f"Validation Passed : {target_table}"
    )
```

```python
validate_table(
    "cicd_dev.gold.gold_customer_analytics",
    "cicd_test.gold.gold_customer_analytics",
    "customer_id"
)

validate_table(
    "cicd_dev.gold.gold_product_analytics",
    "cicd_test.gold.gold_product_analytics",
    "product_id"
)

validate_table(
    "cicd_dev.gold.gold_sales_analytics",
    "cicd_test.gold.gold_sales_analytics",
    "sale_date"
)

validate_table(
    "cicd_dev.gold.gold_inventory_analytics",
    "cicd_test.gold.gold_inventory_analytics",
    "product_id"
)
```
