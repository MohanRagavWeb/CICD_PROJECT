```python
def promote_table(
    source_table,
    target_table
):

    try:

        df = spark.table(source_table)

        record_count = df.count()

        df.write \
          .format("delta") \
          .mode("overwrite") \
          .saveAsTable(target_table)

        print(
            f"SUCCESS: {source_table} -> {target_table}"
        )

        print(
            f"Records Loaded: {record_count}"
        )

    except Exception as e:

        print(
            f"FAILED: {source_table}"
        )

        print(str(e))

        raise
```

```python
promote_table(
    "cicd_test.gold.gold_customer_analytics",
    "cicd_prod.gold.gold_customer_analytics"
)
```

```python
promote_table(
    "cicd_test.gold.gold_product_analytics",
    "cicd_prod.gold.gold_product_analytics"
)
```

```python
promote_table(
    "cicd_test.gold.gold_sales_analytics",
    "cicd_prod.gold.gold_sales_analytics"
)
```

```python
promote_table(
    "cicd_test.gold.gold_inventory_analytics",
    "cicd_prod.gold.gold_inventory_analytics"
)
```
